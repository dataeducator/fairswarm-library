# FairSwarm Quickstart

Select a fair coalition of federated learning clients in under 20 lines of code.

```python
from fairswarm import FairSwarm, FairSwarmConfig, Client, DemographicFitness
from fairswarm.core.client import create_synthetic_clients
from fairswarm.demographics import DemographicDistribution

# 1. Create 20 clients with 4 demographic groups
clients = create_synthetic_clients(n_clients=20, n_demographic_groups=4, seed=42)

# 2. Target: equal representation across all groups
target = DemographicDistribution.uniform(4)

# 3. Set up the optimizer
config = FairSwarmConfig(swarm_size=15, fairness_weight=0.7)
optimizer = FairSwarm(clients=clients, coalition_size=8, config=config, target_distribution=target, seed=42)

# 4. Run it
fitness_fn = DemographicFitness(target_distribution=target)
result = optimizer.optimize(fitness_fn=fitness_fn, n_iterations=50)

# 5. Inspect
print(f"Coalition: {[clients[i].id for i in result.coalition]}")
print(f"Fitness:   {result.fitness:.4f}")
print(f"DemDiv:    {result.fairness.demographic_divergence:.6f}")
print(f"Fair:      {result.fairness.epsilon_satisfied}")
```
